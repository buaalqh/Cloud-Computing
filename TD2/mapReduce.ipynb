# First Part : Introduction 

MapReduce is a programming model useful in for big data processing and allows running a parallel and distributed  algorithm on a cluster.

A MapReduce algorithm is composed of:
<ul>
    <li> a map function, which performs a first set of operations on the input data and, usually, produces a set of key-value pairs as output  </li>
    <li> a reduce function, which performs an operation on the data coming as output from the map function grouped on the basis of the key.</li>
</ul>

The MapReduce algorithm can be run on a framework (i.e. Hadoop). In this case the framework will handle the process of
running the algorithm on the distributed architecture running the various tasks in parallel, managing all communications and data transfers between the various parts of the system, and providing for redundancy and fault tolerance. The map function and the reduce function will then become independent procedures running in multiple instances in parallel.

The map and the reduce procedures can be seen then as two main functions: map() et reduce().

### Map

In general the map() function is seen as a function that:
<ul>
    <li> takes as input one (or a series) key/value pair(s);  </li>
    <li> processes each key/value pair according to the procedure it implements;</li>
    <li> generates as output zero or more key/value pairs.</li>
</ul>

It is important to notice that:
<ul>
    <li> if the input data is composed by single values each sigle value can be seen as a key/value pair having an empty  value;  </li>
    <li> the types of the input key/value pair and the output key/value pair(s) can be (and often are) different from each other;</li>
    <li> output key/value pair can have a dummy key;</li>
    <li> also the output key/value pair can have an empty value.</li>
</ul>

In the example shown in the slides the application is counting the number of occurrences of each word in the provided file. The map function would take as input a single line. This single line can be seen as a key/value pair having as key the whole line and as value an empty value. The function will break the line into words and output a key/value pair for each word. The key will be the string representing the word and the value will be the integer 1. 


### Shuffle
Between the map() and reduce() functions, the data are shuffled in order to move together data sharing the same key.
Data will be then processed by the reduce function. 
When you run a MapReduce procedure on a framework the shuffle operation is fully and transparently handled by the framework. 




### Reduce
The reduce() function applies a procedure on all the pairs that share the same key.  

It is important to notice that:
<ul>
    <li> the reduce function produces zero or more outputs for each group of key-value pairs  </li>
    <li> the type of the output can be different from the input and can be also one or a set of key/value pair(s);</li>
    <li> even in this case the key/value pair can have an empty value.</li>
</ul>



In the word count example, the reduce function takes the input values, sums them, and generates a single output containing a set of key/value pairs having as key the value of each word and as value the final sum.

## Step 1 - map()


<p align="justify">
<font size="3">
We can encode each line of the document as a list of integers.
The map function will then take as input this list of integers and we are going to define a Python function that generates, given a list of integer, a pair (o,1) for each odd number and a pair (e,1) for each even number.  

Suppose that the file contains the following lines:

1 2 3 3

3 4 4

5 6 7

9 8 7 4

In this example for the first line we will have

(o,1), (e,1), (o,1), (o,1)

In order to read the line you can use the function array_reader that takes as input a file name and reads the file:

```python
def array_reader(filename):
    numbers = [line.strip() for line in open(filename)]
    newnumbers = []
    for ln in numbers :
        newnumbers.append([int(i) for i in ln.split(" ")])
    print(newnumbers)
    return (newnumbers)
```

```python
#Provide here few comment lines that describe your map function

def map_odd_even_sol(array):
    pairs = []
    #to complete with your code
    for line in array :
        buffer = []
        for e in line :
            if int(e) % 2 == 0 :
                buffer.append(('e',1))
            else:
                buffer.append(('o',1))
        pairs.append(buffer)
        
    return (pairs)

array_numbers = array_reader("numbers.txt")
res = map_odd_even_sol(array_numbers)
print(res)
```

How can we improve our function?

```python
#Provide here few comment lines that describe your map function

def map_odd_even_simple_sol(array):
    res = []
    #to complete with your code
    for line in array:
        res.append([('o',1) if (num % 2 ==1) else('e',1) for num in line])
    return res
        
array_numbers = array_reader("numbers.txt")
res = map_odd_even_simple_sol(array_numbers)
print(res)
```

## Step 2 - shuffle()

<p align="justify">
<font size="3">
Here we are going to define a functions that groups the pairs according to the values of the key.

We have a list of list of pairs:

[[(o, 1), (e, 1), (o,1), (e,1)], [(o, 1),(e,1),(e,1)], [(o, 1), (e, 1), (o,1)], [(o, 1), (e, 1), (o,1), (e,1)]]

and we have to create two lists of pairs:

[[(o, 1), (o, 1), (o, 1), (o, 1), (o, 1), (o, 1), (o, 1)], [(e, 1), (e, 1), (e, 1), (e, 1), (e, 1), (e, 1), (e, 1)]]

</font>
</p>



```python
#Provide here few comment lines that describe your shuffle function

def shuffle_odd_even_sol(pairs):
    odd = []
    even = []

    for line in pairs :
        for (k,v) in line :
            if k=='o':
                odd.append((k,v))
            else:
                even.append((k,v))
                
    result = [odd,even] 
    return (result)

res = shuffle_odd_even_sol(map_odd_even_sol(array_numbers))
print(res)
```

## Step 3 - reduce()

<p align="justify">
<font size="3">
Here we are going to define the function that it is going to produce the final result. 
We are going to run the reduce function for each key (here for each different group, then twice).
</font>
</p>


```python
#Provide here few comment lines that describe your reduce function

def reduce_sol(pairs):
    #to complete with your code
    result = []
    for line in pairs :
        result.append((line[0][0],len(line)))
        
    return result

#Provide here the code that calls your reduce function according to your implementation

res = reduce_sol(shuffle_odd_even_sol(map_odd_even_sol(array_numbers)))
print(res)
    
```

```python
import json 
def readData(filename):
    with open(filename, mode='r', encoding='utf-8') as file:
        for line in file:
            record = json.loads(line)
            yield(record)
print(list(readData('books.json')))
```

# Exercise 4. Matrix Multiplication

<p>
<font size="3">
We want to define the MapReduce procedure to obtain the matrix $A \times B$. Here you are not suppose to write code but just to think about the representation of the two matrix and to define the general description of the two procedures.
</font>
</p>

<p>
<font size="3">
We have as input:
<ul>
<li> $A$: an $n \times m$ matrix
<li> $B$: an $m \times p$ matrix.
</ul>
The procedure returns the matrix obtained by multiplying the first and the second matrix.
The multiplication can only be computed if the number of columns of $A$ equals the number of rows of $B$.
</font>
    
</p>


<p>
Consider the following A matrix:
<p>
<center>
  $A= \begin{bmatrix}
    1 & 2 & 4   \\
    4 & 5 & 6 
\end{bmatrix}$
</center>
</p>

<p>
Then consider the following B matrix:
<p>
<center>
  $B= \begin{bmatrix}
    a & b    \\
    c & d   \\
    e & f 
\end{bmatrix}$
</center>
</p>

<p>
<font size="3">
Since we are in a MapReduce context can represent these two matrix as a set of key-value pairs, where the key is the coordinate (row identifier, column identifier) of an element, and the value is the element itself.

This matrix representation is the starting point for defining your matrix representation in the following part of the exercise.
</font>
</p>

<p>
<font size="3">
For the matrix A: $( (0, 0), 1 ), ( (0, 1), 2 ), ( (0, 2), 4 ), ( (1, 0), 4 ), ( (1, 1), 5 ), ( (1, 2), 6 )$
    
For the matrix B: $( (0, 0), a ), ( (0, 1), b ), ( (1, 0), c ), ( (1, 1), d ), ( (2, 0), e ), ( (2, 1), f )$

</font>
</p>


<p>
<font size="3">
The result of the matrix multiplication will be a matrix $C = A \times B$ is a $n \times p$ matrix, where each element $c_{i, k}$ is computed as 
follows:
<center>
  $c_{i, k} = \sum\limits_{j=0}^{m-1} a_{i, j} \cdot b_{j, k} \quad\quad (1)$ 
</center>
</font>
</p>

<p>
<center>
  $C= \begin{bmatrix}
    1a + 2c + 4e & 1b + 2d + 4f   \\
    4a + 5c + 6e & 4b + 5d + 6f   \\
\end{bmatrix}$
</center>
</p>



<p>
<font size="3">

As you can see, for a given $(i, k)$, the element of $A$ that is in the j-th column is multiplied by the value of $B$ that is in the j-th row, for any column $j$.
   
    
    
    
For example the element C(0,0) = A(0,0) B(0,0) + A(0,1) B(1,0) + A(0,2) B(2,0) where $j = 0$
</font>
</p>

<p>
<font size="3">
Therefore, we can change the representation of $A$ and $B$ so that their elements are indexed by using $j$ as the key. The $j$ will be used to correctly group the elements to run the map and reduce functions.
    
Remember that the element C(0,0) is computed according to $j=0$.
</font>
</p>

<p>
<font size="3">
More specifically, we can represent the matrix $A$ as follows:
<center>    
    $(j, (0, i, a_{i, j})) \quad 0 \leq i \leq n-1 \quad 0 \leq j \leq m-1  \quad\quad (2)$
</center>    
Since we must still be able to identify to which input matrix each value belongs to, we must add a marker in the representation. In this example we choose the value $0$ in the triple $(0, i, a_{i, j})$ meaning that the element $a_{i, j}$ comes from the matrix $A$. 
    
Notice that in this example we can think that $(j, (0, i, a_{i, j}))$ is a key-value pair where the key is $j$ and the value is $(0, i, a_{i, j})$
</font>
</p>

<p>
<font size="3">
For the matrix A we will then have: $( 0, (0, 0, 1 )), ( 1, (0, 0, 2 )), ( 2, (0, 0, 4 )), ( 0, (0, 1, 4 )), ( 1, (0, 1, 5 ), ( 2, (0, 1, 6 ))$
</font>
</p>

<p>
<font size="3">
Similarly, we can represent the matrix $B$ as follows:
<center>    
    $(j, (1, k, b_{j, k})) \quad 0 \leq j \leq m-1 \quad 0 \leq k \leq p-1 \quad\quad (3) $
</center>    
even in this case the value $1$ in the triple $(1, k, b_{j, k})$ means that the element $b_{j, k}$ comes from the matrix $B$.
</font>
</p>

<p>
<font size="3">
For the matrix B we will then have: $( 0, (1, 0, a )), ( 0, (1, 1, b )), ( 1, (1, 0, c )), ( 1, (1, 1, d )), ( 2, (1, 0, e ), ( 2, (1, 1, f ))$
</font>
</p>

<p>
<font size="3">
Now we can group the values obtained above by their key $j$. 
    
We will have:
<ol>
    <li> ( 0, (0, 0, 1 )), ( 0, (0, 1, 4 )), ( 0, (1, 0, a )), ( 0, (1, 1, b ))</li>

<p>
<font size="3">
As we have seen each element can be seen as a key-value pair, where the key is the index $j$ and the value is a triple corresponding to the elements of matrix $A$ in the $j-$th column  and the elements of matrix $B$  in the $j-$row, as follows:

<p>
Remember that all triples corresponding to matrix $A$ have 0 as their first value, while those corresponding to matrix $B$ have 1.
</p>
<p>
From Equation (1), you can see that the product $a_{i, j} \cdot b_{j, k}$ contributes to the value $c_{i, k}$, for $0 \leq i \leq n-1$ and $0 \leq k \leq p-1$.
